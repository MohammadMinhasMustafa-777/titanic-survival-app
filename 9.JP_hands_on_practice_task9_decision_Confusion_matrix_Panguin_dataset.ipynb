```python
import seaborn as sns
import joblib
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
from matplotlib import pyplot as plt
```

```python
df = sns.load_dataset('Penguins')
dataset = df.dropna()
X = dataset[['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm', 'body_mass_g']]
y = dataset['species']


print(dataset)

saved_model = joblib.load('Best_model_Penguin.pkl')
print("File as been loaded! ", saved_model)
```

```python
pred = saved_model.predict(X)
cm = confusion_matrix(y, pred)
disp = ConfusionMatrixDisplay(cm, display_labels = ['Gentoo', 'Adelie', 'Chinstarp']).plot(cmap= 'Blues')
plt.title("How the data in dataset Varies")
plt.show()
```
